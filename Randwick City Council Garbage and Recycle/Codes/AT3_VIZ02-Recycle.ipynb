```python
import pandas as pd
import folium
```

```python
data = pd.read_csv(r'D:\UTS\Coursework\Semester 1\Data Visualisation and Narrative\AT3\Data\My cleaning\recycleAvgerage.csv')
data.head()
```

```python
m = folium.Map(location = [-33.914589, 151.241714],  tiles="CartoDB positron", zoom_start = 13)
m
```

```python
borderstyle = {
    "color": "black",
    "weight": 2,
    "fill": False
}

folium.GeoJson(data = r'D:\UTS\Coursework\Semester 1\Data Visualisation and Narrative\AT3\Data\geojson\randwick_geojson.json',
              name = "Borders",
              style_function = lambda x: borderstyle).add_to(m)


rubbishlayer  = folium.FeatureGroup('Garbage').add_to(m)
for itr in range(len(data)):
    latstr = data.iloc[itr]["latitude"]
    lngstr = data.iloc[itr]["longitude"]
    capstr = data.iloc[itr]["weight"]
    streetstr = data.iloc[itr]["street_new"]
    print_string = 'Street: {}, Weight: {}'.format(streetstr,capstr)
    clr = 'green'
    cradius = capstr*0.8
    folium.Circle(
        location = [latstr,lngstr],
        radius = cradius,
        popup =print_string,
        color = clr,
        weight = 0.5,
        fill = True
    ).add_to(rubbishlayer)


folium.Circle(
        location = [-33.9517296, 151.2462459], # FOR NOW -  MANUAL, CAN BE MADE DYNAMIC,
        radius = 33.8*0.7,
        color = '#ff1aff',
        weight = 0.7,
        fill = True
    ).add_to(m)


folium.LayerControl().add_to(m)
m
```
