```python
import pandas as pd
```

```python
data = pd.read_csv(r'D:\UTS\Coursework\Semester 1\Data Visualisation and Narrative\AT3\Data\Cleaning\mapped.csv')
data.head(100)
```

```python
gdata = data[data.garbage_type == 'recycle']
```

```python
gdata.head(100)
```

```python
gdata.shape
```

```python
ggdata = gdata.groupby(by = "street_new")["weight"].mean()
```

```python
ggdata.to_csv('recycleAverage.csv')
```

```python
gdata = gdata.sort_values(by = "street_new", axis = 0)
gdata.head()
```

```python
locdf = gdata[['street_new', 'latitude', 'longitude']]
```

```python
locdf.head()
```

```python
locdf = locdf.drop_duplicates()
```

```python
locdf.shape
```

```python
ggdata.shape
```

```python
recycleAvg = pd.merge(ggdata,locdf, on = "street_new")
```
